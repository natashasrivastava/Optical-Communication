```python
import matplotlib.pyplot as plt
import math
```

```python
ricore = 1.48
ricladdingstart  = 1.46
ricladdingend = 1.43
ricladding = []
count = ricladdingstart
while(count>=ricladdingend):
    ricladding.append(float(count))
    count = count-0.0005

NA = []
ricoreSquare = ricore ** ricore
for i in ricladding:
    NA.append(((ricoreSquare) - (i**i))**0.5)
print(NA)

```

```python
AcceptanceAngle = []
for i in NA:
    AcceptanceAngle.append(math.asin(i))
print(AcceptanceAngle)
```

```python
AcceptanceAnglewater =[]
for i in NA:
    AcceptanceAnglewater.append(math.asin(i)/1.33)
print(AcceptanceAnglewater)
```

```python
plt.plot(ricladding,NA, label='Numerical Aperture',marker='o', markerfacecolor='blue', markersize=3)
plt.plot(ricladding,AcceptanceAngle,label='Acceptance Angle',marker='o', markerfacecolor='blue', markersize=3)
plt.plot(ricladding,AcceptanceAnglewater,label='Acceptance Angle Water',marker='o', markerfacecolor='blue', markersize=3)
plt.xlabel('RI of Cladding')
plt.ylabel('Numerical Apperture')

plt.title('w.r.t cladding RI')

plt.legend()

plt.show()
```
