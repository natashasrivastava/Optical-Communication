```python
import matplotlib.pyplot as plt
import math
```

```python
ricladding = 1.46
ricore =[]
ricoreStart  = 1.48
ricoreEnd = 1.52
count = ricoreStart
while(count<=ricoreEnd):
    ricore.append(float(count))
    count = count+0.0005

NA = []
ricladdingSquare = ricladding ** ricladding
for i in ricore:
    NA.append(((i**i) - (ricladdingSquare))**0.5)
print(NA)
```

```python
AcceptanceAngle = []
for i in NA:
    AcceptanceAngle.append(math.asin(i))
print(AcceptanceAngle)
```

```python
AcceptanceAnglewater =[]
for i in NA:
    AcceptanceAnglewater.append(math.asin(i)/1.33)
print(AcceptanceAnglewater)
```

```python
plt.plot(ricore,NA, label='w.r.t cladding RI',marker='o', markerfacecolor='blue', markersize=3)
plt.plot(ricore,AcceptanceAngle,label='w.r.t cladding RI',marker='o', markerfacecolor='blue', markersize=3)
plt.plot(ricore,AcceptanceAnglewater,label='Acceptance Angle Water',marker='o', markerfacecolor='blue', markersize=3)
plt.xlabel('RI of Core')
plt.ylabel('Numerical Apperture')

plt.title('Answer 1 w.r.t. core RI')

plt.legend()

plt.show()
```
